```python
import numpy as np
import config as c
import matplotlib.pyplot as plt
from assim_tools import *
from parallel import *
import sys, os, time
```

```python
c.comm = parallel_start()
c.nproc = 8
c.comm.size = c.nproc
```

```python
state_info = parse_state_info(c)
obs_info = parse_obs_info(c)
```

```python
obs_inds = np.load('/cluster/work/users/yingyue/obs_inds.1.npy', allow_pickle=True).item()
```

```python
nfld = len(state_info['fields'])
nens = c.nens
ny, nx = c.grid.x.shape
```

```python
fld_list = [(m, f) for m in range(nens) for f in range(nfld)]
fld_list_proc = assim_tools.distribute_tasks(comm, fld_list)
```

```python
np.max([len(fld_lst) for pid,fld_lst in fld_list_proc.items()])
```

```python
nx_tile = np.minimum(20, int(np.round(np.sqrt(c.nx*c.ny/c.nproc/3))))
#nx_tile = int(np.round(c.dx_tile/c.dx))
tile_list = [(i, np.minimum(i+nx_tile, c.nx), j, np.minimum(j+nx_tile, c.ny))
             for j in np.arange(0, c.ny, nx_tile) for i in np.arange(0, c.nx, nx_tile)]
mask = c.mask
nloc_tile = np.array([np.sum((~mask[j1:j2, i1:i2]).astype(int)) for i1,i2,j1,j2 in tile_list])
nobs_tile = 0
load_on_tile = np.maximum(nloc_tile, 1) * np.maximum(nobs_tile, 1)
tile_list_proc = distribute_tasks(c.comm, np.arange(len(tile_list)), load_on_tile)
```

```python
##show load on map
fig, ax = plt.subplots(1, 1, figsize=(8,8))

c.grid.plot_land(ax)

cmap = [plt.cm.jet(x) for x in np.linspace(0, 1, c.nproc+1)]

x_ = c.grid.x[0, :]
y_ = c.grid.y[:, 0]
for p_id, t_id_list in tile_list_proc.items():
    for t_id in t_id_list:
        i1,i2,j1,j2 = tile_list[t_id]
        obs_count = np.sum([len(obs_inds[r][t_id]) for r in obs_info['records'].keys()])
        ax.fill([x_[i1], x_[i1], x_[i2-1], x_[i2-1], x_[i1]], [y_[j1], y_[j2-1], y_[j2-1], y_[j1], y_[j1]], color=cmap[p_id], alpha=.5)
        ax.text(x_[i1], y_[j1], '{}'.format(obs_count), fontsize=5)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,8))

c.grid.plot_land(ax, showgrid=False)

for proc_id in range(c.nproc):
    for i1,i2,j1,j2 in tile_list_proc[proc_id]:
        ii_tile, jj_tile = np.meshgrid(np.arange(i1,i2), np.arange(j1,j2))
        ind = ~mask[j1:j2, i1:i2]
        ax.scatter(x_[ii_tile[ind]], y_[jj_tile[ind]], 3, color=cmap[proc_id])
```
