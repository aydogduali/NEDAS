```python
import numpy as np
import matplotlib.pyplot as plt
from models import nextsim
```

```python
filename = "/cluster/work/users/yingyue/data/nextsim_ens/001/field_20070110T000000Z.bin"
seaice_thick = nextsim.read_data(filename, 'M_thick')
seaice_velocity = nextsim.read_data(filename, 'M_VT').reshape((2,-1))  ##velocity u, v are stored in one variable
grid = nextsim.get_grid(filename)
```

```python
##visualise the data
fig, ax = plt.subplots(1, 1, figsize=(7,6))
grid.plot_field(ax, seaice_thick, vmin=0, vmax=3)  ##ice thickness in shading
grid.plot_vectors(ax, seaice_velocity, linecolor='w')  ##vectors showing the ice motion
grid.plot_land(ax, color='gray') ##show the map
```

```python
##zoom in on the Fram Strait and show the triangular mesh
fig, ax = plt.subplots(1, 1, figsize=(7,7))
ax.tripcolor(grid.tri, seaice_thick, vmin=0, vmax=3, cmap='viridis')
ax.triplot(grid.tri, color='w', linewidth=0.2)
grid.plot_land(ax, color='gray')
ax.set_xlim(0, 1.2e6)
ax.set_ylim(-1.2e6, 0)
```
