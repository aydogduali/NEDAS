```python
import numpy as np
import matplotlib.pyplot as plt
from models import topaz
```

```python
f = topaz.abfile.ABFileBathy('/mnt/nird/TP4/test/depth_TP4b0.12_01.a', 'r', idm=800, jdm=880)
mask = f.read_field('depth').mask
```

```python
filename = "/mnt/nird/TP4/test/TP4restart2007_002_00_mem001.a"
```

```python
u = topaz.read_data(filename, 'u', 1, mask)
v = topaz.read_data(filename, 'v', 1, mask)
current = np.array([u, v])
ssh  = topaz.read_data(filename, 'ssh', 0, mask)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,9))

topaz.grid.plot_field(ax, ssh, vmin=-0.8, vmax=0.8, cmap='bwr')
topaz.grid.plot_vectors(ax, current, V=0.1, num_steps=50)
topaz.grid.plot_land(ax, color=[.7, .7, .5], dlon=20, dlat=20)
```
